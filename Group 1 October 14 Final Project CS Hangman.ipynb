```python
###This is the whole coded corrected for HANGMAN to run properly - Ampra
import random

Drawingframe = ["""
        _______
        |     |
        |     
        |
        |
        |
        |
        |___________
        |SAVE A LIFE|""", """

        _______
        |     |
        |     O
        |
        |
        |
        |
        |__________
        |TRY HARDER|""", """


        _______
        |     |
        |     O
        |   --|
        |
        |
        |
        |__________________________
        |I STILL KNOW YOU CAN DO IT|""", """


        _______
        |     |
        |     O
        |   --|--
        |
        |
        |
        |__________________________________
        |IT'S NOT EASY FOR YOU BUT TRY AGAIN|""", """


        _______
        |     |
        |     O
        |   --|--
        |     |
        |    /
        |
        |___________
        |LAST CHANCE|""", """


        _______
        |     |
        |    *_*
        |    /|\\
        |     |
        |    / \\
        |
        |________
        |GAME OVER|"""]

def gameStart():
    
    print("Welcome to Two-Player Hangman.")


def getRandomWord():
    word_category = input("Please choose one of these word categories (Please copy the spelling exactly as shown): Fruit, Animal, Car Brand, Country. ")

    wordsFruit = ['apple', 'banana', 'mango', 'strawberry', 'orange', 'grape', 'pineapple', 'lemon', 'coconut', 'watermelon', 'cherry', 'papaya', 'berry', 'peach', 'lychee']
    wordsAnimal = ['antelope', 'bird', 'cat', 'crocodile', 'dog', 'dolphin', 'fish', 'horse', 'kangaroo','leopard','monkey','panda', 'pig', 'rat', 'wolf']
    wordsCarBrand = ['chevrolet', 'honda', 'nissan', 'ford', 'fiat', 'jeep', 'volkswagen', 'volvo', 'jaguar', 'audi', 'toyota', 'porsche', 'bugatti', 'bentley', 'tesla']
    wordsCountry = ['canada','australia','albania','bangladesh','bhutan','bolivia','chile','denmark','egypt','finland','hungary','iceland','italy','mexico','mongolia']

    if word_category == 'Fruit': 
      word = random.choice(wordsFruit)
      return word
    
    elif word_category == 'Animal': 
      word = random.choice(wordsAnimal)
      return word

    elif word_category == 'Car Brand': 
      word = random.choice(wordsCarBrand)
      return word

    elif word_category == 'Country':
      word = random.choice(wordsCountry)
      return word

    else:
      print("Invalid category option. Ending game.")
      exit()



Wrongguesses = ''
Rightanswers = ''
secretWord = getRandomWord()
gameIsDone = False

# Defining Functions and Parameters in this code chain
# "secretWord" is the random word identifier 

def Outputframe(Drawingframe, Wrongguesses, Rightanswers, secretWord):
    print(Drawingframe[len(Wrongguesses)])
    print()

    print('Missed Letters:', end=' ')
    for letter in Wrongguesses:
        print(letter, end=' ')
    print("\n")
    

blanks = '_' * len(secretWord)

# This replaces blanks with correctly guessed letters

for i in range(len(secretWord)):
    if secretWord[i] in Rightanswers:
        blanks = blanks[:i] + secretWord[i] + blanks[i+1:]

# This shows the secret word with spaces in between each letter

    for letter in blanks:  
        print(letter, end=' ')
    print("\n")

# This code chain defines functions for errors like when more than one word is inputted by the player or if the input is not a an alphabetical letter  

def getGuess(alreadyGuessed):
    while True:
        guess = input('Guess a letter for player ones word: ')
        guess = guess.lower()
        if len(guess) != 1:
            print('Please enter a single letter for player ones word.')
        elif guess in alreadyGuessed:
            print('You have already guessed that letter for player ones word. Choose again.')
        elif guess not in 'abcdefghijklmnopqrstuvwxyz':
            print('Please enter a single letter for player ones word.')
        else:
            return guess

# This part of the code enables the definition to have another playing match between player 1 and player 2
def playAgain():
    return input("\nDo you want to try again player 2? Just Type yes or no - ").lower().startswith('y')



while True:
    Outputframe(Drawingframe, Wrongguesses, Rightanswers, secretWord)

    guess = getGuess(Wrongguesses + Rightanswers)

    if guess in secretWord:
        Rightanswers = Rightanswers + guess

        foundAllLetters = True
        for i in range(len(secretWord)):
            if secretWord[i] not in Rightanswers:
                foundAllLetters = False
                break
        if foundAllLetters:
           print("\n")
           print("---- SCORE BOARD ----")
           print("\n")
           print("            The secret word  - " +  secretWord)
           print("            Number Of Wins   - " + str(len(Rightanswers)))
           print("            Number Of Losses - " + str(len(Wrongguesses)))
           print("            Status           - You have won! ")

           gameIsDone = True
    else:
        Wrongguesses = Wrongguesses + guess

        if len(Wrongguesses) == len(Drawingframe) - 1:
            Outputframe(Drawingframe, Wrongguesses,
                         Rightanswers, secretWord)
            print('You have run out of guesses!')
            print("\n")
            print("---- SCORE BOARD ----")
            print("\n")
            print("            The secret word  - " +  secretWord)
            print("            Number Of Wins   - " + str(len(Rightanswers)))
            print("            Number Of Losses - " + str(len(Wrongguesses)))
            print("            Status           - You have lost! ")
            gameIsDone = True

    if gameIsDone:
        if playAgain():
            Wrongguesses = ''
            Rightanswers = ''
            gameIsDone = False
            secretWord = getRandomWord()
        else:
            break


print("\n")
print("---- SCORE BOARD ----")
print("\n")
print("            The secret word  - " +  secretWord)
print("            Number Of Wins   - " + str(len(Rightanswers)))
print("            Number Of Losses - " + str(len(Wrongguesses)))
print("            Status           - You have won! ")
```
